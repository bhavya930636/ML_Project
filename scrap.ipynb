```python
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

with open('/home/sjain/Documents/sem4/DL,ML/catch/links', 'r') as links_file:
    lines = links_file.readlines()



for line in lines:

    case_name, link = line.strip().split(',')
    
    print(case_name)
    print(link)
    folder_path = f'/home/sjain/Documents/catch/D2V-BiGRU-CRF/citations/{case_name}/'

    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    driver = webdriver.Chrome()
    driver.get(link)

    covers_div = driver.find_element(By.CLASS_NAME, "covers")
    citetop_span = covers_div.find_element(By.CLASS_NAME, "citetop")
    first_hyperlink = citetop_span.find_element(By.TAG_NAME, "a")
    first_hyperlink_url = first_hyperlink.get_attribute("href")
    
    driver.get(first_hyperlink_url)
    time.sleep(2)
    import re

    def sanitize_title(title):
        # Replace special characters with underscores
        sanitized_title = re.sub(r'[^\w\s-]', '_', title)
        # Remove leading and trailing whitespace
        sanitized_title = sanitized_title.strip()
        # Replace consecutive spaces with single space
        sanitized_title = re.sub(r'\s+', ' ', sanitized_title)
        # Replace spaces with underscores
        sanitized_title = sanitized_title.replace(' ', '_')
        return sanitized_title
    
    


    # Function to process the current page
    def process_page(driver, folder_path):
        result_title_divs = WebDriverWait(driver, 5).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "result_title"))
        )
        t = 0
        i=0
        
        for result_title_div in result_title_divs:
            result_title_text = result_title_div.text
            print(result_title_text)
            i+=1
            print(i)
            if 'vs' in result_title_text.lower():
                print("VS")
                
                hyperlink = result_title_div.find_element(By.TAG_NAME, "a")
                href_link = hyperlink.get_attribute("href")
                print(href_link)
                sanitized_title = sanitize_title(result_title_text)
                print(sanitized_title)
                file_name = f'{sanitized_title}.txt'
                
                file_path = os.path.join(folder_path, file_name)

                with open(file_path, 'w+', encoding='utf-8') as file:
                    print("File_opening")
                    driver.get(href_link)
                    time.sleep(2)  # Optional: Wait for the page to load
                    judgments_element = driver.find_element(By.CLASS_NAME, 'judgments')
                    judgments_content = judgments_element.text
                    # Write the content to the file
                    file.write(judgments_content)
                    
                driver.back()
                

            else:
                print("else")
                hyperlink = result_title_div.find_element(By.TAG_NAME, "a")
                href_link = hyperlink.get_attribute("href")
                print(href_link)
                sanitized_title = sanitize_title(result_title_text)
                print(sanitized_title)
                file_name = f'{sanitized_title}.txt'
                file_path = os.path.join(folder_path, file_name)

                while True:
                    try:
                        with open(file_path, 'w+', encoding='utf-8') as file:
                            print("File_opening")
                            driver.get(href_link)
                            time.sleep(1)
                            act_element = driver.find_element(By.CLASS_NAME, 'akoma-ntoso')
                            act_content = act_element.text
                            # Write the content to the file
                            file.write(act_content)
                        break  # Exit the loop if file writing is successful
                    except FileNotFoundError:
                        # Handle file not found error by creating the file and retrying
                        os.makedirs(os.path.dirname(file_path), exist_ok=True)
                        continue
                    except Exception as e:
                        print(f"An error occurred: {e}")
                        break  
                driver.back()

            
    # Process the current page
    process_page(driver, folder_path)

    # Check if there's a "Next" button and process the next page if it exists
    next_button = driver.find_elements(By.LINK_TEXT, "Next")
    if next_button:
        next_button[0].click()
        print("next page")
        process_page(driver, folder_path)

    driver.quit()
```
